```python
#Salary parsing
salary = df['Salary Estimate'].apply(lambda x: x.split("(")[0])
```

```python
print(salary)
```

```python
minus_kd = salary.apply(lambda x: x.replace("K"," ").replace("$"," "))
```

```python
print(minus_kd)
```

```python
df['min_salary'] = minus_kd.apply(lambda x: x.split("-")[0])
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce')
```

```python
df['min_salary']
```

```python
print(df)
```

```python
df['max_salary'] = minus_kd.apply(lambda x: x.split("-")[1])
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')
#df['max_salary'].dtype
```

```python
print(df['max_salary'])
```

```python
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
print(df['avg_salary'])
```

```python
df = df[df['Competitors'] != '-1']
print(df['Competitors'])
```

```python
# Assuming 'Company Name' column contains strings like "ABC Corporation\, Inc."
df['company_text'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
```

```python
#sate field
# Function to split location by comma and check number of elements
def count_split_elements(location):
    # Split location by comma (',') to separate parts
    parts = location.split(',')
    # Return the number of elements produced after splitting
    return len(parts)

# Apply the function to create a new column 'num_elements'
df['num_elements'] = df['Location'].apply(lambda x: count_split_elements(x))

# Function to extract the second part of the location if it has multiple elements
def extract_second_part(location):
    # Split location by comma (',') to separate parts
    parts = location.split(',')
    # Check if there are at least two elements after splitting
    if len(parts) > 1:
        # Return the second part (index 1) after stripping whitespace
        return parts[1].strip()
    else:
        # Return None if there is only one element (no second part)
        return None

# Apply the function to create a new column 'second_part'
df['job_state'] = df['Location'].apply(lambda x: extract_second_part(x))

# Print the DataFrame to view the results
print(df['job_state'])
```

```python
df.job_state.value_counts()
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters
 else 0, axis =1)
print(df['same_state'])
```

```python
#age of company
df = df[df['Founded'] != -1]
df['age'] = df.Founded.apply(lambda x: x if x < 1 else 2024 - x)
print(df['age'])
```
